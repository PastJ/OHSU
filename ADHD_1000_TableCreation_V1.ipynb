```python
def find_columns_with_string(df, search_string):
    columns_with_string = []
    for column in df.columns:
        if search_string in column:
            columns_with_string.append(column)
    return columns_with_string
```

```python
# Create a new table that merges: (Pollutants + Lead Risk + FARA + ADI > GeoCoded)
GeoCoded = ['Pollutants', 'Lead Risk', 'FARA', 'ADI']

# Change the selected measures above to GeoCoded
filtered_data_dict['Table'] = filtered_data_dict['Table'].apply(lambda x: 'GeoCoded' if x in GeoCoded else x)
```

```python
# All unique tables 
tables = filtered_data_dict["Table"].unique()
tables
```

```python
base_file_path  = r"D:\OHSU\TAN_Project\Data\ABCD\SI\ADHD_1000\Table_Year_Dropped"
for table in tables:
    years = filtered_data_dict[filtered_data_dict["Table"] == table]["Year"].unique() # data_dict_df
    df = filtered_data_dict[filtered_data_dict["Table"] == table] # data_dict_df
    
    for year in years:
        # Fetch and clean the 'Description' values where 'Year' is 'Y0'
        vars = list(df[df["Year"] == year]["Description"].values)

        # Remove spaces and $ at the end of strings
        vars = [s.rstrip().rstrip('$') for s in vars]  

        # Convert all column names in the list to uppercase
        vars_upper = [v.upper() for v in vars]
        
        # Check if 'vars_upper' exists in the merged DataFrame
        try:
            df_new = merged_df_updated[["MERGEID"] + vars_upper] # merged_df
            
            # Construct the specific file path
            file_path = os.path.join(base_file_path, f"{table}_{year}.csv")
            
            # Save the DataFrame to the specified location
            df_new.to_csv(file_path, index=False)
            print(f"CSV saved to: {file_path}")
        
        except KeyError as e:
            print(f"Error: {e}. Columns {vars_upper} not found in merged_df_updated.")
```
